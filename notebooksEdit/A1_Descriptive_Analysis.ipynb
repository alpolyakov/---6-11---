<p><b>Задача 2.</b> Анализ рынка акций</p>

<p>1. Загрузите данные в один dataframe из всех файлов в папке /data/stock. Все файлы имеют одинаковую структуру, в том числе наименование столбцов. В качестве значений индекса dataframe'а необходимо указать значения столбца "Date". Название столбцов должные соответствовать названию акций (имя файла без .csv), а их значения - значениям цены закрытия (столбец "Close" в файлах .csv)</p>

```python
import os
filedir = r"C:\Users\Yoda\Desktop\Assignments-master\data\A1_Descriptive_Analysis\stock"
filelist = os.listdir(filedir)
li = []
stockNames = []

for i in filelist:
    dfStock = pnd.read_csv(filedir + '\\' + i, index_col=0, parse_dates=True) # загрузка из файла в dataframe
    li.append(dfStock)
    stockNames.append(i.replace('.csv',''))

frame = pnd.concat(li, axis=0, keys = stockNames, ignore_index=False)
df = frame["Close"]
df
```

<p>2. Рассчитайте корреляционную матрицу для всех акций</p>

```python
df2 = pnd.DataFrame()
for i in stockNames:
    df2[i] = df[i]
corrMatrix = df2.corr()
corrMatrix
```

<p>3. Отобразите корреляционную матрицу в виде диаграммы. Примерный вид графика приведен ниже</p>
<img src="img/fig_corr_matrix.png" width="400px">

```python
fig, ax = plt.subplots(figsize = (12,8)) 

c = ax.matshow(corrMatrix, vmin=-1, vmax=1)
fig.colorbar(c, ax = ax) 
ax.set_title('Корреляционная матрица акций\n\n')

ax.set_xticklabels(stockNames)
ax.set_xticks(np.arange(len(stockNames)))
plt.setp(ax.get_xticklabels(), rotation=90, ha="left",
         rotation_mode="anchor")

ax.set_yticklabels(stockNames)
ax.set_yticks(np.arange(len(stockNames)))

plt.show()
```
